```python
# 1. How many months are there on Mars?
df.month.nunique()
```

```python
# 2. How many sols (Martian days) worth of data are there?
df.sol.count()
```

```python
# 3. What is the average minimum temperature by month?
df.groupby("month").min_temp.mean()
```

```python
# Plot the average minimum temperature by month
df.groupby("month").min_temp.mean().plot(kind="bar", title="Average Minimum Temperature by Month", ylabel="Temperature", color="darkred")
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph
df.groupby("month").min_temp.mean().sort_values().plot(kind="bar", title="Average Minimum  Temperature by Month", ylabel="Temperature", color="darkred")
plt.show()
```

```python
# 4. What is the average pressure by month?
df.groupby("pressure").min_temp.mean()
```

```python
# Plot the average pressure by month
df.groupby("month").pressure.mean().plot(kind="bar", title="Average Pressure by Month", ylabel="Pressure", color="darkgreen")
plt.show()
```

```python
# Identify the lowest and highest pressure months in Curiosity's location by sorting the previous graph
df.groupby("month").pressure.mean().sort_values().plot(kind="bar", title="Average Pressure by Month", ylabel="Pressure", color="darkgreen")
plt.show()
```

```python
# 5. How many terrestrial (Earth) days are there in a Martian year?
# Visually estimate the result by plotting the daily minimum temperature of each observation in the data set.
# Change Canvas Size
plt.figure(figsize=(12, 6))

# Create a basic plot
plt.plot(df.terrestrial_date, df.min_temp, color="darkred")

# Add Labels/Title
plt.xlabel("Terrestrial Date", fontfamily="Arial", fontsize=12)
plt.ylabel("Minimum  Temperature", fontfamily="Arial", fontsize=12)
plt.title("Minimum  Temperature by Day", fontweight="bold", fontsize=16, fontfamily="Arial")

# Add in a background grid
plt.grid(axis="y", linestyle="--", color="lightgrey", alpha=0.75)

# Show/Save the Graph
plt.show()
```
